```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
plt.title('Species Count')
sns.countplot(iris['species'])
```

```python
plt.plot(iris["sepal_length"], "r--")
plt.show()
```

```python
iris.plot(kind = 'scatter', x = 'sepal_length', y = 'petal_width')
plt.grid()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, size = 5)
plt.show()
```

```python
iris.plot(kind="scatter",x="sepal_length",y="sepal_width",color="green",s=60)
plt.show()
```

```python
sns.violinplot(x="species", y="petal_length",data=iris,size=8)
plt.show()
```

```python
sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris, size = 5, ratio = 10, kind = 'hist', color = 'red' )
plt.show()
```

```python
sns.boxplot(x="species",y= "petal_length",data=iris)
plt.show()
```

```python
sns.pairplot(iris,hue="species",height=4)
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",height=5)\
.map(sns.distplot,"petal_length")\
.add_legend()
plt.show()
```
